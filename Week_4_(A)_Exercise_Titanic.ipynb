#### 8. Create three unique <b>pivot tables</b> using the titanic data. Make one statement about something you can determine from the information each table shows.

<b>EXAMPLE:</b> "The table above shows that employees use their vacation leave most frequently in December, June, and August."

```python
pd.pivot_table(titanic, index = ['pclass'], values = ['fare'], columns = ['survived'])

# did people who pay more fare have a higher chance of survival?
# 0 is people who did not survive, 1 is people who survived: the average fare that people paid in each category is shown
```

```python
pd.pivot_table(titanic, index = ['pclass', 'sex'], values = ['survived'], aggfunc = ['count'])

# How many people from each sex in each pclass survive?
```

```python
pd.pivot_table(titanic, index = ['home.dest'], values = ['body'])
```

```python
pd.pivot_table(titanic, index = ['pclass', 'survived'], values = ['age'])

# what is the passaenger average age based on the ticket class and chance survival?
```

```python
pd.pivot_table(titanic, index = ['boat'], values = ['survived'], aggfunc = ['count', np.sum, np.mean])

# how many people were assigned to each bot and did they all survive?
# 1.0000 means that 100% of the people assigned to that boat survived
# On boat 12, 19 people were assigned and only 18 people survived so 94.7% of them survived
```
