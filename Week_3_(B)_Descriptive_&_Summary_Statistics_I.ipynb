## Pivot Tables

Pandas pivot tables allow you to get another point of view of your data with only one line of code. By default, the pandas pivot tables will show you the mean of all numeric columns in your dataset - grouped by a specific index category. However, if you want more control, you can input the specific parameters that you want to see.

The basic pivot table code is simple:

<b>pd.pivot_table ( dataset name, index=[ grouping variable ], values=[ column to aggregate ] )</b>

- dataset name is self explanatory - it's the DataFrame you'd like to use
- index is the column (or list of columns) you'd like to group your data by. It will be displayed in the index column (or columns, if you're passing in a list)
- values (optional) is the column you'd like to aggregate. If you do not specify this then the function will aggregate all numeric columns.

```python
# basic pivot table - by default, the aggregate variables will display the 'mean'
# will aggregate all numeric columns

pd.pivot_table(df, index=["Type 1"])
```

```python
# basic pivot table - by default, the aggregate variables will display the 'mean'
# will aggregate only the attack column

pd.pivot_table(df, index=["Type 1"], values=["Attack"])
```

```python
# mean for each values of the 'Name' column
a = pd.pivot_table(df, index=['Name'], values = ['Speed'])
a.loc['Ivysaur']
```

```python
# basic pivot table - by default, the aggregate variables will display the 'mean'
# will aggregate only the attack column
# output is further grouped by the stage column

pd.pivot_table(df, index=["Type 1"], values=["Attack"], columns=["Stage"])
```

## Pivot Tables -- Summary Statistics

Until now, our pivot tables have only shown the average value for our specified column. But what if we want to see operations besides the mean? We can specify a few simple operations with the <b>aggfunc</b> parameter.

- aggfunc (optional) accepts a function or list of functions you'd like to use on your group (default: numpy.mean).

The following options are possible when using the aggfunc column:

- np.mean = average
- np.median = median
- np.std = standard deviation
- np.sum = sum of values
- 'count' = count of observations
- min = minimum value
- max = maximum value

You can specify one specific operation OR you can pass a list with multiple operations.

```python
## include the aggfunc = to specify the types of output you want to see

pd.pivot_table(df, index=['Type 1'], values=["Attack"], aggfunc= [np.mean, np.median, min, max, np.std,'count', np.sum])
```

## Grouping Data by Conditions

Similar to using pivot tables, grouping your data in different ways can give you insight to the characteristics of the dataset. You can use the following functions to group your data: mean, max, min, sum, std, median

When you group your dataset by a variable - it makes the most sense to group by a 'categorical' variable. However, the function will work with both categorical and numeric variables.

```python
# group your entire dataset by the 'Stage' column
# output the mean for each numeric variable by stage

df.groupby(df['Stage']).mean()
```

```python
# group the 'Defense' variable by the "Stage" variable
# output the mean defense for each stage

df["Defense"].groupby(df['Stage']).mean()
```

```python
# group the 'Defense' variable by the "Type 1" variable
# output the minimum defense for each pokemon type

df["Defense"].groupby(df['Type 1']).min()
```

```python
# group both the "Defense" and "Attack" variables by "Stage"
# output the average defense and attack for each stage

df[['Defense', 'Attack']].groupby(df['Stage']).mean()
```

# Finding Data Irregularities

Missing and duplicate data are two of the most common issues you will encounter when exploring your dataset. It is important to identify how significant either of these issues are before you attempt ANY additional analyses. Next class we will talk about how to handle missing or duplicate data, but for now, we will focus on how to simply identify these issues.

## Missing Data

```python
# Missing Data - does my dataset have ANY missing data?
# the output will show how many 'missing' values are within that specific column

df.isnull().sum()
```

```python
# What if you want to identify the specific rows that are missing for a specific column?
# locating rows with missing values

# create a dataset that shows where Type 2 is missing;
# this will show up as a (true/false) list
df_missing = df["Type 2"].isnull()

# output the first few rows of this dataset
df_missing.head()
```

```python
# next, use the .loc function to output the instances where df_missing == True

df.loc[df_missing]
```

## Duplicate Data

```python
# bringing in a dataset with known duplicate data

dupes = pd.read_csv("pokemon_dupe.csv")
dupes.head()
```

```python
# use the .duplicated() function to check the entire dataset for duplicate rows (ALL values must match to be considered a duplicate)
# create a dataset that shows the rows where the data is duplicated; this will show up as a boolen (true/false) list

df_dupes = dupes.duplicated()

# check the first few rows of this dataset
df_dupes.head()
```

```python
# next use the .loc function to output the rows where dupes == True

dupes.loc[df_dupes]
```
