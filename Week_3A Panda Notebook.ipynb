```python
poke = pd.read_csv('pokemon.csv', index_col = "Name")
poke.head(12)
#index_col is now changed from default to the names of the pokemons
```

```python
poke.loc["Pikachu"]
#don't need row or column numbers, just need the names of the cell
```

```python
poke.loc[['Metapod', "Weedle", 'Charmander', 'Mew']]
#selecting multiple rows with loc
```

```python
poke.loc[["Metapod", 'Weedle', 'Charmander', 'Mew'], 'Type 1']
#multiple non-consecutive rows with 1 column
```

```python
poke.loc[['Metapod', 'Weedle', 'Charmander', 'Mew'], ['Type 1', 'Stage', 'Legendary']]
#show multiple non-consecutive rows and columns
```

```python
poke.loc['Bulbasaur':'Squirtle']
#range of rows with no restriction on columns
```

```python
poke.loc['Bulbasaur': 'Squirtle', ['Type 1', 'Stage']]
#range of rows and specific columns
```
